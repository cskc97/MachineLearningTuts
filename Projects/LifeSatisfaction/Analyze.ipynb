```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sb
import sklearn
import sklearn.linear_model
import sklearn.preprocessing
from sklearn.preprocessing import scale

```

```python
h_index = pd.read_csv("HappinessIndex2016.csv")
h_index.head()
```

```python
part_h = h_index.loc[:,["Happiness Score","Freedom","Economy (GDP per Capita)"]]
```

```python
part_h.head()
```

```python
sb.pairplot(part_h)
```

```python
happiness = part_h["Happiness Score"].values.reshape(-1,1)
freedom = part_h["Freedom"].values.reshape(-1,1)
```

```python
model = sklearn.linear_model.LinearRegression(normalize=True)
model.fit(freedom,happiness)
```

```python
model.score(freedom,happiness)
```
