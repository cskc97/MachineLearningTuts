```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import sklearn
import sklearn.linear_model
%matplotlib inline
```

```python
happiness_index_csv = pd.read_csv("HappinessIndex2016.csv")
happiness_index_csv

```

```python
#obtaining the models

happiness_score = happiness_index_csv["Happiness Score"]
gdp_per_capita = happiness_index_csv["Economy (GDP per Capita)"]
```

```python
plt.plot(gdp_per_capita,happiness_score)
plt.show()
```

```python
#Creating a model - Linear Regression observed from the graph of the data above.
model = sklearn.linear_model.LinearRegression()
model.fit(gdp_per_capita.values.reshape(1,-1),happiness_score.values.reshape(1,-1)) #reshaping done just for the data to work with the sklearn module
```

```python
make_prediction_for_X_value = 1.52000
print(model.predict(make_prediction_for_X_value.values.reshape(1,-1)))
```
